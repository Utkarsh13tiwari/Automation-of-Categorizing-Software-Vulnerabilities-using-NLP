```python
text = " ".join(review for review in df["Threat"])
print ("There are {} words in the combination of all Threat.".format(len(text)))
```

```python
stopwords = set(STOPWORDS)
```

```python
stopwords.update(["id", "from", "could", "use", "also","Many","like","May","may","ca","even","ect","i.e","cause","result","used","form","Its","cause","take","following","vulnerability"])
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
